```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'])
df['shipping date (DateOrders)'] = pd.to_datetime(df['shipping date (DateOrders)'])

# Calculate actual shipping duration
df['Actual Shipping Time'] = (df['shipping date (DateOrders)'] - df['order date (DateOrders)']).dt.days
df['Days for shipment (scheduled)'] = pd.to_numeric(df['Days for shipment (scheduled)'], errors='coerce')
df['Schedule Accuracy'] = df['Days for shipment (scheduled)'] - df['Actual Shipping Time']
```

```python
# Distribution of Actual Shipping Time
plt.figure(figsize=(8,5))
sns.histplot(df['Actual Shipping Time'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Actual Shipping Time")
plt.xlabel("Days between Order and Shipping")
plt.ylabel("Frequency")
plt.show()
# Shows how shipping time is distributed overall, helping identify if most orders are shipped quickly or experience delays.
```

```python
# Shipping Time by Shipping Mode 
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Shipping Mode', y='Actual Shipping Time')
plt.title("Shipping Time by Shipping Mode")
plt.xlabel("Shipping Mode")
plt.ylabel("Actual Shipping Time (days)")
plt.show()
# Compares average and spread of shipping time across different shipping modes (e.g., Standard, First Class) to find the most reliable method
```

```python
# Shipping Time by Region
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Order Region', y='Actual Shipping Time', palette='coolwarm')
plt.title("Shipping Time by Region")
plt.xticks(rotation=45)
plt.show()
# Identifies regional differences in shipping performance to spot areas experiencing longer shipping times
```

```python
# Products with Longest Average Shipping Time
product_avg_shipping = df.groupby('Product Name')['Actual Shipping Time'].mean().sort_values(ascending=False).head()

plt.figure(figsize=(10,6))
sns.barplot(x=product_avg_shipping.values, y=product_avg_shipping.index, palette="Blues")
plt.title("Products with Longest Average Shipping Time")
plt.xlabel("Average Shipping Time (days)")
plt.ylabel("Product Name")
plt.show()
# Highlights which products take the longest to ship, helping in optimizing product-specific logistics.
```

```python
# Schedule Compliance Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Schedule Accuracy'], bins=20, kde=True, color='skyblue')
plt.title("Shipping Schedule Accuracy")
plt.xlabel("Scheduled - Actual (days)")
plt.axvline(0, color='red', linestyle='--', label='On Schedule')
plt.legend()
plt.show()
# Visualizes the difference between scheduled and actual shipping times. Negative values = delays, Positive = early/on-time.
```

```python
# Departments with Worst Schedule Accuracy
dept_delay = df.groupby('Department Name')['Schedule Accuracy'].mean().sort_values()

plt.figure(figsize=(10,6))
sns.barplot(x=dept_delay.values, y=dept_delay.index, palette='coolwarm')
plt.title("Department Average Schedule Compliance")
plt.xlabel("Avg. Schedule Accuracy (days)")
plt.ylabel("Department")
plt.show()
# Identifies which departments have the worst average delays compared to scheduled shipping dates.
```

```python
# High-Risk Deliveries by Customer Segment
segment_risk = df[df['Late_delivery_risk'] == 1]['Customer Segment'].value_counts()

plt.figure(figsize=(10,6))
sns.barplot(x=segment_risk.values, y=segment_risk.index, palette='Set1')
plt.title("High-Risk Deliveries by Customer Segment")
plt.xlabel("Count")
plt.ylabel("Customer Segment")
plt.show()
# Displays which customer segments are most affected by high-risk deliveries, helping in targeted customer experience improvements.
```

```python
# Delivery Timeliness Categories
df['Delivery Status Category'] = df['Schedule Accuracy'].apply(
    lambda x: 'Early' if x > 0 else 'On Time' if x == 0 else 'Late')

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Delivery Status Category', order=['Early', 'On Time', 'Late'], palette='Set2')
plt.title("Delivery Timeliness Distribution")
plt.xlabel("Delivery Status")
plt.ylabel("Number of Orders")
plt.show()
# Shows the proportion of early, on-time, and late deliveries to evaluate overall delivery performance
```

```python
# Delivery Status by Shipping Mode
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Shipping Mode', hue='Delivery Status Category', palette='Set3')
plt.title("Delivery Timeliness by Shipping Mode")
plt.xlabel("Shipping Mode")
plt.ylabel("Number of Orders")
plt.legend(title='Delivery Status')
plt.show()
# Helps identify which shipping mode performs most reliably in terms of timeliness
```
